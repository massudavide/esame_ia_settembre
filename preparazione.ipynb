```python
import numpy as np
import pandas as pd
```

```python
path = "heart_failure_clinical_records_dataset.train.csv"
data = pd.read_csv(path)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split

# seleziono la feature di output... si molto inutile
y = data["DEATH_EVENT"]
# seleziono le features che compariranno nel DataSet
X = data

# divido i dati in train (70%) e test (30%) e li mischio in modo da averli non ordinati,
# ma in modo deterministico (random_State = 42)  
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
# salvo i file in csv per utilizzarli nell'altro notebook
train_X.to_csv('train.csv')
test_X.to_csv('test.csv')
```
